# Data import

```python
import yfinance as yf

df = yf.Ticker('0005.HK').history(period='10y')
df
```

```python
df = df.filter(["Close"])
df = df.rename(columns={"Close": "GT"})
df
```

# Data Visualize - 日K圖

```python
import matplotlib.pyplot as plt
#import matplotlib.style as mpl

#plt.style.use('classic') 
plt.style.use("seaborn-v0_8")   # plt.style.use("seaborn-darkgrid"x is deprecated 
plt.xlabel("Date")
plt.ylabel("price")
plt.plot(df["GT"],linewidth = 1)
plt.show
```

# Standardize data (a.k.a. Normalization)

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
scaled_prices = scaler.fit_transform(df.values)
scaled_prices
```

# Data Construction

```python
import numpy as np

MOVING_WIN_SIZE = 60

all_x, all_y = [], [] # ==> format in python list, need to alter to numpy narray format
for i in range(len(scaled_prices) - MOVING_WIN_SIZE):
    x = scaled_prices[i:i+MOVING_WIN_SIZE]
    y = scaled_prices[i+MOVING_WIN_SIZE]
    all_x.append(x)
    all_y.append(y)

all_x, all_y = np.array(all_x), np.array(all_y)
print(f"x.shape:{all_x.shape}\ny.shape:{all_y.shape}")
```

# Spliting Data

```python
split_rate = 0.8

train_data_size = round(all_x.shape[0] * split_rate)
train_x, train_y = all_x[:train_data_size], all_y[:train_data_size]
test_x, test_y = all_x[train_data_size:], all_y[train_data_size:]
print(f"train_x.shape:\t{train_x.shape}\ttrain_y.shape:\t{train_y.shape}")
print(f"test_x.shape:\t{test_x.shape}\ttest_y.shape:\t{test_y.shape}")
```

# Construct model

```python
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.models import Sequential

# RNN structure = LSTM*2 + Dense + output, 1st LSTM layer use to identify micro-features and 2nd LSTM layer use to identify macro-features

model = Sequential()
model.add(LSTM(units = 50, return_sequences = True, input_shape = (train_x.shape[1], 1)))
model.add(LSTM(units = 50, return_sequences = False))
model.add(Dense(units = 25))
model.add(Dense(units = 1))

model.summary()
```

if LSTM layer set too much units which increases many neurons, it results in the issue of overfitting 

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

# Verify Hardware Acceleration

```python
#Mac only - testing GPU operation
import tensorflow as tf

# Print TensorFlow version
print("TensorFlow version:", tf.__version__)

# List physical devices
physical_devices = tf.config.list_physical_devices()
print("Physical devices:", physical_devices)

# List physical devices again
gpu_devices = tf.config.list_physical_devices('GPU')
print("GPUs:", gpu_devices)

# Simple matrix multiplication to test GPU
with tf.device('/GPU:0'):
    a = tf.constant([[1.0, 2.0], [3.0, 4.0]])
    b = tf.constant([[1.0, 1.0], [0.0, 1.0]])
    c = tf.matmul(a, b)
    print("Matrix multiplication result:\n", c)
```

# Train Model

```python
from tensorflow.keras.callbacks import EarlyStopping

callback = EarlyStopping(monitor = 'val_loss', patience = 10, restore_best_weights = True)
model.fit(train_x,train_y,
         validation_split = 0.2,  # take the 20% data from the end to act as validation data set
         callbacks = [callback],
         epochs = 1000)
         
```

## result explaination

Loss drop with the number of epoch increase for the train data set. Which mean the accuracy is increasing only to the traning set while the over-fitting getting worse.

In order to prevent serious over-fitting issue. Validation data was used

The validation loss and loss descend at the beginning of the epoch. The loss will keep decrease while the validation loss rise after the certain epoch which indicate the drawback of the over-fitting float out

The **best result** is the **lowest validation loss**

# Evaluate Model

```python
preds = model.predict(test_x)
preds
```

```python
preds_unscaled = scaler.inverse_transform(preds)
preds_unscaled
```

# Visualise Result

```python
train_df = df[:train_data_size + MOVING_WIN_SIZE]
test_df = df[train_data_size + MOVING_WIN_SIZE:]
test_df = test_df.assign(Predict = preds_unscaled)

plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(train_df['GT'], linewidth = 2)
plt.plot(test_df["GT"], linewidth = 2)
plt.plot(test_df["Predict"], linewidth = 1)
plt.legend(["train", "GT", "Predict"])
plt.show()
```

# Problem / Issue of Model

```python
plt.plot(train_df['GT'][-20:], linewidth = 2)
plt.plot(test_df['GT'][:30], linewidth = 2)
plt.plot(test_df['Predict'][:30], linewidth = 1)
plt.legend(["train", "GT", "Predict"])
plt.show()
```

```python
test_df = test_df.assign(Shifted = test_df['GT'].shift(1))
test_df.iat[0,-1] = train_df.iat[-1,-1]
test_df
```

## Terminology alert: CvRMSE

CvRMSE = Coefficient of Variation of the Root Mean Squared Error

```python
from sklearn.metrics import root_mean_squared_error

predict_rmse = root_mean_squared_error(test_df['GT'], test_df['Predict'])
predict_cvrmse = predict_rmse / test_df['GT'].mean() * 100
predict_cvrmse
```

```python
shifted_rmse = root_mean_squared_error(test_df['GT'], test_df['Shifted'])
shifted_cvrmse = shifted_rmse / test_df['GT'].mean() * 100
shifted_cvrmse
```

```python
if predict_cvrmse < shifted_cvrmse:
    print("Prdicted result is better")
else:
    print("Copying the last day 'close price' is better result than the model prediction")
```
