````python
import os
print(os.path.exists('./data/product_sales_2023.csv'))
print(os.listdir('./data'))
````

````python
# import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# read data
df = pd.read_csv('./data/product_sales_2023.csv')
df['date'] = pd.to_datetime(df['date'])

# add month column
df['month'] = df['date'].dt.to_period('M')

# monthly sales
monthly_sales = df.groupby('month')['units_sold'].sum()
monthly_sales.plot(kind='bar', figsize=(12, 5), title='Monthly Units Sold')
plt.xticks(rotation=45)
plt.show()
````

````python
# mirror copying last graph => not a good presentation

monthly_sales_per_region = df.groupby(['month', 'region'])['units_sold'].sum()
monthly_sales_per_region.plot(kind='bar', figsize=(12, 5), title ='Monthly Units Sold Per Region')
plt.xticks(rotation=45)
plt.show()
````

````python
region_monthly = df.groupby(['month', 'region'])['units_sold'].sum().unstack()
region_monthly.plot(kind='line', figsize=(12,6), title='Monthly Units Sold by Region')
````

````python
grouped = df.groupby(['month', 'region'])['units_sold'].sum()

#把 region 拉成橫向欄位（方便畫圖）
pivot = grouped.unstack()
````

````python
customer_grouped = df.groupby(['month', 'customer_type'])['units_sold'].sum()

#把 region 拉成橫向欄位（方便畫圖）
pivot = customer_grouped.unstack()
pivot.plot(kind='line', figsize=(12, 6), title='Monthly Units Sold by Customer Type')
````

````python
type_monthly = df.groupby(['month', 'customer_type'])['units_sold'].sum().unstack()
type_monthly.plot(kind='line', figsize=(12,6), title='Monthly Units Sold by Customer Type')
````

````python
df.groupby(['month', 'region'])['price_per_unit'].mean().unstack().plot()
````

````python
monthly_price = df.groupby('month')['price_per_unit'].mean()
monthly_price.plot(kind='line', figsize=(12, 5), title='Monthly Avg Price per Unit')
````

````python
avg_price = df.groupby(['month', 'region'])['price_per_unit'].mean().unstack()
avg_price.plot(kind='line', figsize=(12, 6), title='Monthly Avg Price per Unit by Region')
plt.ylabel('Avg Price per Unit')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
df.groupby('month')['price_per_unit'].mean().plot(kind='line', marker='o', figsize=(10,5), title='Overall Monthly Avg Price')
plt.ylabel('Avg Price per Unit')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
df.groupby('month')[['price_per_unit', 'units_sold']].mean().plot.scatter(
    x='price_per_unit', y='units_sold', title='Avg Price vs Units Sold'
)
````

````python
correlation = df.groupby('month')[['price_per_unit', 'units_sold']].mean().corr()
print(correlation)
````

````python
df['quarter'] = df['date'].dt.to_period('Q')
df.groupby('quarter')['price_per_unit'].mean().plot(kind='bar', title='Avg Price per Unit by Quarter')
````
