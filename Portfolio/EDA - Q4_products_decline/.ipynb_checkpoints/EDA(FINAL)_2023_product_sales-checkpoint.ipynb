# üß≠ ÂàÜÊûêÁõÆÊ®ô
Êú¨Ê¨°ÂàÜÊûêÁõÆÁöÑÊòØÈáùÂ∞ç 2023 Âπ¥ÂÖ®Âπ¥ÁöÑÁî¢ÂìÅÈä∑ÂîÆË≥áÊñôÔºåÊâæÂá∫Á¨¨ÂõõÂ≠£ (Q4) Èä∑ÈáèÊòéÈ°Ø‰∏ãÈôçÁöÑÂéüÂõ†„ÄÇ

ÊàëÂÄëÂ∞áÂæû‰ª•‰∏ãÈù¢ÂêëÊé¢Ë®éÔºö
- Âú∞ÂçÄÂà•Èä∑ÂîÆÁãÄÊ≥Å
- ÂÆ¢Êà∂È°ûÂûãÁöÑÁµêÊßãËÆäÂåñ
- ÂñÆÂÉπËÆäÂãïËàáÈä∑Èáè‰πãÈñìÁöÑÈóúËÅØÊÄß
- ÊòØÂê¶Â≠òÂú®ÊΩõÂú®ÁöÑÂ∏ÇÂ†¥Á≠ñÁï•ÊîπÈÄ≤Á©∫Èñì


# üßπ Ë≥áÊñôÊ∫ñÂÇôËàáÂü∫Êú¨Ê∏ÖÊ¥ó

```python
# ËºâÂÖ•Â•ó‰ª∂
import pandas as pd
import matplotlib.pyplot as plt

# ËºâÂÖ•Ë≥áÊñô
df = pd.read_csv('./data/product_sales_2023_with_product.csv', parse_dates=['date'])

# Êñ∞Â¢ûÊúà‰ªΩËàáÂ≠£Â∫¶Ê¨Ñ‰Ωç
df['month'] = df['date'].dt.month
df['quarter'] = df['date'].dt.to_period('Q')

# Á∞°ÂñÆÁúãÁúãË≥áÊñô
df.head()
```

# üîé Step 1ÔºöQ4 Èä∑ÂîÆ‰∏ãÊªëÁ¢∫Ë™ç

```python
# ÊåâÂ≠£Â∫¶ÂΩôÁ∏ΩÈä∑ÂîÆÈáè
quarterly_sales = df.groupby('quarter')['units_sold'].sum()

# Áï´Âá∫Â≠£Â∫¶Á∏ΩÈä∑Èáè
quarterly_sales.plot(kind='bar', title='Total Units Sold by Quarter', ylabel='Units Sold', xlabel='Quarter', color='skyblue')
plt.grid(axis='y')
plt.show()
```

## ‚úçÔ∏è Step 1.5ÔºöÂØ´Â∞èÁµêË´ñÔºàÂ∞èÂ∞è‰∏ÄÊÆµÔºâ
### Step 1 Â∞èÁµêË´ñÔºöQ4 Èä∑ÂîÆ‰∏ãÈôçÁ¢∫Ë™ç
ÂæûÂ≠£Â∫¶Á∏ΩÈä∑ÈáèÁöÑÁõ¥Ê¢ùÂúñÂèØ‰ª•ÁúãÂá∫Ôºå2023 Âπ¥Á¨¨ÂõõÂ≠£ (Q4) ÁöÑÈä∑ÈáèÊòéÈ°Ø‰ΩéÊñºÂâç‰∏âÂ≠£„ÄÇÈÄôË≠âÂØ¶‰∫ÜÂàùÊ≠•Êá∑ÁñëÔºåÂç≥ Q4 ÁöÑÈä∑ÂîÆË°®ÁèæÁï∞Â∏∏ÔºåÈúÄË¶ÅÈÄ≤‰∏ÄÊ≠•Êé¢Á©∂ÂéüÂõ†„ÄÇ


# üó∫Ô∏è Step 2ÔºöÂæûÂú∞ÂçÄÂàÜÊûêÈä∑ÂîÆËÆäÂåñ

```python
# ÊåâÂ≠£Â∫¶ËàáÂú∞ÂçÄÂΩôÁ∏ΩÈä∑ÂîÆÈáè
region_quarterly_sales = df.groupby(['quarter', 'region'])['units_sold'].sum().unstack()

# Áï´Âá∫ÊØèÂÄãÂú∞ÂçÄÊØèÂ≠£ÁöÑÈä∑ÈáèËÆäÂåñ
region_quarterly_sales.plot(kind='bar', stacked=False, title='Units Sold by Region and Quarter', figsize=(10, 6))
plt.ylabel('Units Sold')
plt.xlabel('Quarter')
plt.grid(axis='y')
plt.legend(title='Region')
plt.show()
```

## Step 2 Â∞èÁµêË´ñÔºöÂú∞ÂçÄÂàÜÊûê
ÂæûÂú∞ÂçÄËàáÂ≠£Â∫¶Èä∑ÈáèËÆäÂåñ‰æÜÁúãÔºåÂõõÂÄãÂú∞ÂçÄÂú®Á¨¨ÂõõÂ≠£ (Q4) ÁöÑÈä∑ÈáèÂùáÊúâ‰∏ãÈôçÔºå‰∏î‰∏ãÈôçÂπÖÂ∫¶Áõ∏Ëøë„ÄÇÂõ†Ê≠§ÔºåÂú∞ÂçÄÂ∑ÆÁï∞‰∏¶‰∏çÊòØ Q4 Êï¥È´îÈä∑Èáè‰∏ãÈôçÁöÑ‰∏ªË¶ÅÂéüÂõ†„ÄÇ


# üë• Step 3ÔºöÂæûÂÆ¢Êà∂È°ûÂûãÂàÜÊûêÈä∑ÂîÆËÆäÂåñ

```python
# ÊåâÂ≠£Â∫¶ËàáÂÆ¢Êà∂È°ûÂûãÂΩôÁ∏ΩÈä∑ÂîÆÈáè
customer_quarterly_sales = df.groupby(['quarter', 'customer_type'])['units_sold'].sum().unstack()

# Áï´Âá∫ÊØèÂÄãÂÆ¢Êà∂È°ûÂûãÊØèÂ≠£ÁöÑÈä∑ÈáèËÆäÂåñ
customer_quarterly_sales.plot(kind='bar', stacked=False, title='Units Sold by Customer Type and Quarter', figsize=(10, 6))
plt.ylabel('Units Sold')
plt.xlabel('Quarter')
plt.grid(axis='y')
plt.legend(title='Customer Type')
plt.show()
```

## Step 3 Â∞èÁµêË´ñÔºöÂÆ¢Êà∂È°ûÂûãÂàÜÊûê
ÂæûÂÆ¢Êà∂È°ûÂûã (Êñ∞ÂÆ¢Êà∂ËàáÂõûË≥ºÂÆ¢Êà∂) ÁöÑÂ≠£Â∫¶Èä∑ÈáèËÆäÂåñ‰æÜÁúãÔºåÂÖ©ËÄÖÂú®Á¨¨ÂõõÂ≠£ (Q4) ÁöÑ‰∏ãÈôçÂπÖÂ∫¶Áõ∏‰ººÔºå‰∏îÂõûË≥ºÂÆ¢Êà∂‰ªçÁÑ∂Âç†Â§ßÂ§öÊï∏„ÄÇÊú™ÁôºÁèæÂõ†ÁâπÂÆöÂÆ¢Êà∂È°ûÂûãÊµÅÂ§±ËÄåÂ∞éËá¥ÁöÑÁï∞Â∏∏„ÄÇÂõ†Ê≠§ÔºåÂÆ¢Êà∂È°ûÂûãËÆäÂãï‰πü‰∏çÊòØ Q4 Èä∑ÂîÆ‰∏ãÈôçÁöÑ‰∏ªË¶ÅÂéüÂõ†„ÄÇ

# üí∞ Step 4ÔºöÊ™¢Êü•ÂñÆÂÉπËÆäÂåñÔºàprice_per_unitÔºâ

```python
# ÊåâÂ≠£Â∫¶Ë®àÁÆóÂπ≥ÂùáÂñÆÂÉπ
quarterly_price = df.groupby('quarter')['price_per_unit'].mean()

# Áï´Âá∫ÂñÆÂÉπËÆäÂåñ
quarterly_price.plot(kind='line', marker='o', title='Average Price per Unit by Quarter', figsize=(8, 5))
plt.ylabel('Average Price per Unit')
plt.xlabel('Quarter')
plt.grid(True)
plt.show()
```

## Step 4 Â∞èÁµêË´ñÔºöÂñÆÂÉπËÆäÂåñÂàÜÊûê
ÂæûÂñÆÂÉπËÆäÂåñ‰æÜÁúãÔºåÂâç‰∏âÂ≠£ (Q1ÔΩûQ3) ÁöÑÂπ≥ÂùáÂñÆÂÉπËÆäÂãïÂπÖÂ∫¶ÂæàÂ∞èÔºåËÄåÁ¨¨ÂõõÂ≠£ (Q4) ÁöÑÂπ≥ÂùáÂñÆÂÉπÂèçËÄåÂ§ßÂπÖ‰∏ãÈôç„ÄÇÁÑ∂ËÄåÔºåÂñÆÂÉπ‰∏ãÈôç‰∏¶Êú™Â∏∂ÂãïÈä∑ÂîÆÈáèÂ¢ûÂä†ÔºåÂèçËÄåÂú®Q4Âá∫ÁèæÈä∑ÂîÆ‰∏ãÊªëÁöÑÊÉÖÊ≥Å„ÄÇÈÄôË™™ÊòéÂÉπÊ†ºËÆäÂãï‰∏¶ÈùûÈä∑Èáè‰∏ãÈôçÁöÑÁõ¥Êé•ÂéüÂõ†ÔºåÂèØËÉΩÂ≠òÂú®ÂÖ∂‰ªñÂïèÈ°åÈúÄË¶ÅÈÄ≤‰∏ÄÊ≠•ÂàÜÊûê„ÄÇ


# üì¶ Step 5-1ÔºöÊåâÁî¢ÂìÅÈ°ûÂà•ÂíåÂ≠£Â∫¶ÂΩôÁ∏ΩÈä∑Èáè

```python
# ÊåâÂ≠£Â∫¶ËàáÁî¢ÂìÅÈ°ûÂà•ÂΩôÁ∏ΩÈä∑ÂîÆÈáè
product_quarterly_sales = df.groupby(['quarter', 'product'])['units_sold'].sum().unstack()

# Áï´Âá∫ÊØèÂÄãÁî¢ÂìÅÊØèÂ≠£ÁöÑÈä∑ÈáèËÆäÂåñ
product_quarterly_sales.plot(kind='bar', stacked=False, title='Units Sold by Product and Quarter', figsize=(12, 6))
plt.ylabel('Units Sold')
plt.xlabel('Quarter')
plt.grid(axis='y')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
