````python
def login(page):
    # Wait for the iframe (hCaptcha) to load
    WebDriverWait(page, 10).until(
        EC.presence_of_element_located((By.TAG_NAME, 'iframe'))
    ) 

    # Locate and fill login information
    username = page.find_element(By.NAME, "username")
    password = page.find_element(By.NAME, "password")
    
    # !! Debug only - Replace with actual input methods
    print("Input account:")
    my_ac = getpass.getpass()
    password.send_keys(my_ac)
    print("Input password:")
    my_pass = getpass.getpass()
    password.send_keys(my_pass)

    # Tick checkboxes
    page.find_element(By.XPATH, '//*[@id="main"]/div/form/div[4]/div/label/span[1]/input').click()
    page.find_element(By.XPATH, '//*[@id="main"]/div/form/div[5]/div/label/span[1]/input').click()

    # Switch to hCaptcha iframe and solve manually
    iframe = page.find_element(By.TAG_NAME, 'iframe')
    page.switch_to.frame(iframe)
    # Wait for iframe checkbox appear
    WebDriverWait(page, 10).until(
        EC.presence_of_element_located((By.ID, "anchor"))
    )
    page.find_element(By.ID, "anchor").click()
    input("Please complete the hCaptcha and press Enter to continue...")
    page.switch_to.default_content()

    # Submit login form
    page.find_element(By.XPATH, '//*[@id="main"]/div/form/div[6]/button').click()

def navigate_to_jobs_page(page):
    # Wait for the home page to load and navigate to job search page
    WebDriverWait(page, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'a[href="/job-search/"]')))
    page.find_element(By.CSS_SELECTOR, 'a[href="/job-search/"]').click()
    
    # Testing only
    #page.find_element(By.CSS_SELECTOR, 'a[href="/job-search/grad"]').click()

def get_job_data(page):
    # Wait until the total items appear
    WebDriverWait(page, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'MuiButtonBase-root.MuiButton-root.MuiButton-contained.MuiButton-containedPrimary.MuiButton-sizeMedium.MuiButton-containedSizeMedium.MuiButton-root.MuiButton-contained.MuiButton-containedPrimary.MuiButton-sizeMedium.MuiButton-containedSizeMedium.css-1dupsu2')))
    total_items = int(page.find_elements(By.CSS_SELECTOR, 'span.MuiTypography-root.MuiTypography-body1.css-194incw')[1].text)
    total_pages = total_items // 10 + 1
    next_page = 1

    while next_page <= total_pages:
        # loop each item within page
        items = WebDriverWait(page, 10).until(EC.visibility_of_all_elements_located((By.XPATH, "//div[@role='listitem']")))
        for item in items:
            item.click()
            #time.sleep(5)
            # !!debug
            extract_and_save_data(page)
        next_page += 1
        go_to_next_page(page, next_page, total_pages)
        time.sleep(5)
        
    '''
    # loop each item within page - only work for current page
    items = WebDriverWait(page, 5).until(EC.visibility_of_all_elements_located((By.XPATH, "//div[@role='listitem']")))
    for item in items:
        item.click()
        #time.sleep(5)
        # !!debug
        extract_and_save_data(page)
    '''


def extract_and_save_data(page):
    title = page.find_element(By.CLASS_NAME, 'MuiTypography-root.MuiTypography-h5.css-hwuvqj').text
    company = page.find_element(By.CLASS_NAME, 'MuiTypography-root.MuiTypography-inherit.MuiLink-root.MuiLink-underlineNone.MuiTypography-root.MuiTypography-body1.css-10yakh6').text
    reference_code = page.find_element(By.CLASS_NAME, 'MuiTypography-root.MuiTypography-subtitle1.css-eqwy4l').text

    posted_date, closing_date = page.find_elements(By.CSS_SELECTOR, 'span.MuiTypography-root.MuiTypography-body1.css-10q65bp')
    posted_date, closing_date = posted_date.text, closing_date.text
    
    divs_mode = page.find_element(By.XPATH, '//div[@class="MuiBox-root css-4deqai"]')
    mode = extract_employment_modes(divs_mode)    

    divs_salary_vacancy_location = page.find_element(By.CLASS_NAME, 'css-85vcpd')
    salary, vacancy, location = extract_salary_vacancy_location(divs_salary_vacancy_location)

    job_description = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "Job Description")]/following-sibling::div').text
    skills_n_requirements = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "Skills & Requirements")]/following-sibling::div').text
    
    graduate_year = page.find_element(By.XPATH, './/h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Graduation Year")]/following-sibling::div').text
    fields_of_study_xpath = './/h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Fields of Study")]/following-sibling::div | .//h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Fields of Study")]/following-sibling::p'
    fields_of_study = page.find_element(By.XPATH, fields_of_study_xpath).text
    education_level = page.find_element(By.XPATH, './/h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Education Level (pursuing / obtained)")]/following-sibling::p').text

    work_experience_year_label = page.find_element(By.XPATH, './/h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Required Level of Work Experience")]')
    work_experience_year = page.execute_script("return arguments[0].nextSibling.nodeValue;", work_experience_year_label).strip()

    about_company = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "About")]/following-sibling::div').text

    #applymethod not neccessarily given in the page
    try:
        apply_method = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "Application Methods")]/following-sibling::p').text
    except NoSuchElementException:
        apply_method = None
        
    address = extract_info_by_icon(page, 'Address Icon')
    email = extract_info_by_icon(page, 'Email Icon')
    phone = extract_info_by_icon(page, 'Phone Icon')
    fax = extract_info_by_icon(page, 'Fax Icon')
    website = extract_info_by_icon(page, 'Website Icon')

    try:
        note = page.find_element(By.XPATH, './/h3[contains(@class, "MuiTypography-root MuiTypography-subtitle1 css-1as8ric") and contains(text(), "Note")]/following-sibling::div').text
    except NoSuchElementException:
        note = None
    
    nature_of_business = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "Nature of Business")]/following-sibling::ul').text
    job_categories = page.find_element(By.XPATH, './/div[contains(@class, "MuiBox-root css-ww7y7v")]/h2[contains(text(), "Job Categories")]/following-sibling::ul').text

    writer.writerow([title, company, reference_code, posted_date, closing_date, mode, 
                     salary, vacancy, location, 
                     job_description, skills_n_requirements, graduate_year, fields_of_study, education_level, work_experience_year,
                     about_company, 
                     apply_method, address, email, phone, fax, website, note, 
                     nature_of_business, job_categories])

def extract_info_by_icon(page, icon_name):
    try:
        icon_xpath = f"//*[name()='svg' and @alt='{icon_name}']/following-sibling::div | //*[name()='svg' and @alt='{icon_name}']/following-sibling::a"
        info_text = page.find_element(By.XPATH, icon_xpath).text.strip()
        return info_text
    except NoSuchElementException:
        return None
        
def extract_salary_vacancy_location(div):
    salary = div.find_element(By.XPATH, ".//p[contains(text(), 'Salary')]/following-sibling::p").text
    vacancy = div.find_element(By.XPATH, ".//p[contains(text(), 'Vacancies')]/following-sibling::p").text
    location = div.find_element(By.XPATH, ".//h6[contains(text(), 'Location')]/following-sibling::p").text
    return salary, vacancy, location

def extract_employment_modes(div):
    try:
        mode_elements = div.find_elements(By.XPATH, './/p[contains(@class, "MuiTypography-root")]')
        all_modes = []
        for mode in mode_elements:
            all_modes.append(mode.text)
        modes = (', '.join(all_modes))
    finally:
        return modes

def go_to_next_page(page, next_page, total_pages):
    try:
        if next_page <= total_pages:
            buttons = page.find_elements(By.CSS_SELECTOR, 'button.MuiButtonBase-root.MuiButton-root')
            for button in buttons:
                if button.text.strip() == str(next_page):
                    button.click()
                    #WebDriverWait(page, 3).until(EC.staleness_of(button))
                    #WebDriverWait(page, 5).until(EC.visibility_of_all_elements_located((By.XPATH, "//div[@role='listitem']")))
                    return True
        return False
    except NoSuchElementException:
        return False

if __name__ == "__main__":
    start=datetime.now()
    driver = webdriver.Chrome()
    driver.get("https://www.jijis.org.hk/site/login")
    
    file_name = f"jijis_jobs_data_{date.today()}(detail).csv"
    file = open(file_name, "w")
    writer = csv.writer(file)
    writer.writerow(["Title", "Company", "Reference Code" ,"Posted Date", "Closing Date", "Employment Mode", 
                     "Salary", "Vacancy", "Location", 
                     "Job Description", "Skills & Requirements", "Graduation Year", "Fields of Study", "Education Level (pursuing / obtained)", "Required Level of Work Experience",
                     "About Company", 
                     "Application Methods", "Address", "Email", "Phone", "Fax", "Website", "Note", 
                     "Nature of Business", 
                     "Job Categories"])
    
    try:
        login(driver)
        navigate_to_jobs_page(driver)
        get_job_data(driver)
    finally:
        file.close()
        print(f"{file_name} is saved.")
        #time.sleep(3)
        #print(driver.page_source)
        driver.quit()
        print("Run time: ", datetime.now()-start)
````

## Challenges faced during data collection

- **Difficult to locate iframe component** （Easy to solve）
>To handle an iframe, especially one containing a hCaptcha on a login page, you'll need to switch the Selenium driver context to the iframe before interacting with elements inside it. Here’s a step-by-step approach:</br>
>1. Identify the iframe:</br>
Use a method like ```find_element_by_tag_name('iframe')``` or ```find_element_by_xpath("//iframe")``` to locate the iframe element.</br></br>
>2. Switch to the iframe:</br>
Use ```driver.switch_to.frame()``` to switch the context to the iframe.</br></br>
>3. Interact with elements within the iframe:</br>
After switching, you can now locate and interact with the hCaptcha or any other elements inside the iframe.</br></br>
>4. Switch back to the main content:</br>
>After interacting with the iframe, use ```driver.switch_to.default_content()``` to go back to the main page content.</br></br>
- **Handling captcha during Login**
 >Automating hCaptcha is challenging because it's specifically designed to detect and block automated interactions. However, you have a couple of options:
  >1. **Manual Intervention** (used in this project)
  >2. Use Third-Party Services
  >3. Fully Automated Option
- **Catch number of total items** *e.g. <span> element*
>can use Selenium's ```By.CSS_SELECTOR``` or ```By.XPATH```</br></br>
>Locate the span element using **CSS selector**:</br>
>```span_element = driver.find_element(By.CSS_SELECTOR, 'span.MuiTypography-root.MuiTypography-body1.css-194incw')```
>
>Locate the span element using **XPath**:</br>
>```span_element = driver.find_element(By.XPATH, '//span[contains(@class, "MuiTypography-root") and contains(@class, "MuiTypography->body1") and contains(@class, "css-194incw")]')```

You can use CSS_SELECTOR to first find elements with a specific class and then filter them based on the text they contain. However, directly combining class selection and text matching in CSS_SELECTOR is a bit tricky because CSS itself doesn't provide a way to match text content directly. Instead, you would need to use a combination of CSS_SELECTOR and Python logic to achieve this.
